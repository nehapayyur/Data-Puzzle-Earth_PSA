# Intro

Here's is the message from Earth:
'I remember that if you sort the countries (or dependencies) [Guinea, Iran, Trinidad And Tobago, Honduras, Lebanon, Ethiopia, Niger, Afghanistan, India, American Samoa, Cuba, Gabon, Nicaragua, Channel Islands, Martinique] by density of population (DESCENDING) and then only look at the first letter of each country, you'll get my message back!! Can you help me please?'

To help our planet, we preloaded world population and geographical data from the website Worldometer (https://worldometers.info/). Use it to do what the earth says and decode the message!

```python
import pandas as pd

countries = ['Guinea', 'Iran', 'Trinidad and Tobago', 'Honduras', 'Lebanon',
             'Ethiopia', 'Niger', 'Afghanistan', 'India', 'American Samoa',
             'Cuba', 'Gabon', 'Nicaragua', 'Channel Islands', 'Martinique']

country_pop = pd.read_csv('https://raw.githubusercontent.com/hyperc54/data-puzzles-assets/master/misc/earth/population_by_country.csv')
```

```python
country_pop.head()
```

```python
country_surface = pd.read_csv('https://raw.githubusercontent.com/hyperc54/data-puzzles-assets/master/misc/earth/surface_by_country.csv')
```

```python
country_surface.head()
```

```python
# Your turn now!
#merge the two dfs
country_combined = country_pop.merge(country_surface[['Land Area (Km²)', 'Country (or dependency)']], on = 'Country (or dependency)', how = 'left')
print(country_combined)
```

```python
#create a new column that contains values of population density
country_combined['Population_density'] = country_combined['Population (2020)']/country_combined['Land Area (Km²)']
```

```python
#filter rows based on only the countries we are interested in given in the 'countries' list
country_interested = country_combined[country_combined['Country (or dependency)'].isin(countries)]
country_interested
```

```python
#sort the df based on descending order of population density
country_sorted = country_interested.sort_values(by = 'Population_density', ascending = False)
country_sorted.head(20)
```

```python
#create new column that contains only first letters of coountry names
country_sorted['new_col'] = country_combined['Country (or dependency)'].astype(str).str[0]
country_sorted
```

```python
#extract the column with first letters of contries into a series
country_series = country_sorted['new_col']
```

```python
#concatenate the country letters into a single string
message = country_series.str.cat(sep='')
```
